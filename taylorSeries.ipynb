```
def func_cos(x,n):
  cos_aprox = 0
  for i in range(n):
    coef=(-1)**i
    num = x**(2*i)
    den = math.factorial(2*i)
    cos_aprox += ( coef ) * ( (num)/(den) )
    print(cos_aprox)
  return cos_aprox
```

```
#Input
ang_rad = (math.radians(int(input("Ingrese el valor que quiere evaluar con la serie de Taylor:"))))
itera = (int(input("Ingrese cuantas iteraciones de la serie quiere hacer:")))
```

```
#Reasignando
out_list=np.zeros(itera)
cos_list=np.zeros(itera)
error_list=np.zeros(itera)

for i in range(itera):
  out = func_cos(ang_rad, i)
  math_cos = math.cos(ang_rad)
  e_error = round((abs(out-math_cos)),15)
  out_list[i]=out
  cos_list[i]=math_cos
  error_list[i]=e_error

fig = go.Figure(data=[go.Table(header=dict(values=['Coseno por serie', 'Coseno por Python', 'Error']),
                 cells=dict(values=[out_list, cos_list, error_list]))])
fig.show()
```

```
#Plot
%matplotlib inline
angles = np.arange(-2*np.pi,2*np.pi,0.1)
p_cos = np.cos(angles)

fig, ax = plt.subplots()
ax.plot(angles,p_cos)

# add lines for between 1 and 6 terms in the Taylor Series
for i in range(1,6):
    t_cos = [func_cos(angle,i) for angle in angles]
    ax.plot(angles,t_cos)

ax.set_ylim([-7,4])

# set up legend
legend_lst = ['cos() function']
for i in range(1,6):
    legend_lst.append(f'Taylor Series - {i} terms')
ax.legend(legend_lst, loc=3)

plt.show()
```
